# [dbo].[isValidPhoneNumber]

```sql
/****** Object:  StoredProcedure [dbo].[InsertCustomer]    Script Date: 27.11.2022 23:42:08 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

create procedure [dbo].[InsertCustomer]
	@customer_name varchar(30),
	@email varchar(30),
	@phone_number varchar(30),
	@address varchar(30),
	@user_password varchar(30)
as
BEGIN
	BEGIN TRANSACTION
		INSERT INTO users(username,
				user_password,
				user_role) 
			VALUES (left(@email, charindex('@', @email) - 1),
					HASHBYTES('SHA2_256', @user_password), 
					'customer')
		IF (@@error <> 0)
        	ROLLBACK
		INSERT INTO customers(user_id, 
				email,
				customer_name, 
				phone_number,
				address)
			VALUES ((select user_id from users
					where username=left(@email,  charindex('@', @email) - 1)),
				@email,
				@customer_name,
				@phone_number,
				@address
			)
		IF (@@error <> 0)
        	ROLLBACK
	COMMIT
END
GO
```
